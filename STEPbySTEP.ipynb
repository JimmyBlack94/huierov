```python
import openai
import os

# Konstanten
MAX_RETRIES = 5

# Daten aus dem Ordner "Große Roadmap" laden
def load_big_roadmap():
    with open(os.path.join("Roadmap Rohling", "roadmap.txt"), "r") as file:
        return file.read()

# Daten aus dem Ordner "Schlüsselmomente" laden
def load_key_moments():
    moments = []
    for filename in os.listdir("Schlüsselmomente"):
        if filename.endswith(".txt"):
            with open(os.path.join("Schlüsselmomente", filename), "r") as file:
                moments.append(file.read())
    return "\n".join(moments)

with open("Complete_Roadmap.txt", "r") as file:
    complete_roadmap = file.read()

big_roadmap = load_big_roadmap()
key_moments = load_key_moments()

# Extrahieren der Roadmap-Abschnitte aus der Complete_Roadmap
def extract_roadmap_sections(complete_roadmap):
    roadmap_parts = complete_roadmap.split('\n')
    total_parts = len(roadmap_parts)

    intro_parts = roadmap_parts[:int(total_parts*0.25)]  # Die ersten 25% der Teile
    middle_parts = roadmap_parts[int(total_parts*0.25):int(total_parts*0.75)]  # Die mittleren 50% der Teile
    ending_parts = roadmap_parts[int(total_parts*0.75):]  # Die letzten 25% der Teile

    roadmap_sections = {
        'Einleitung': ' '.join(intro_parts),
        'Hauptteil': ' '.join(middle_parts),
        'Schluss': ' '.join(ending_parts),
    }

    return roadmap_sections

roadmap_sections = extract_roadmap_sections(complete_roadmap)

# Generierung aller Kapitelintentionen
def generate_all_chapter_intentions():
    intentions = complete_roadmap.split('\n')
    refined_intentions = []
    for intention in intentions:
        keywords = intention.split()[:5]  # Nehmen Sie beispielsweise die ersten 5 Wörter.
        refined_intention = " ".join(keywords)
        refined_intentions.append(refined_intention)
    return refined_intentions

# Kohärenz-Check
def check_coherence(chapter_content, intention):
    keywords = intention.split()
    if not keywords:  # Überprüfung, ob die Liste leer ist
        return False
    present_keywords = [keyword for keyword in keywords if keyword in chapter_content]
    return len(present_keywords) / len(keywords) >= 0.4


# Kapitel Generierung basierend auf der Intention
def generate_chapter_based_on_intention(chapter_num, intention, feedback=None):
    section = 'Einleitung' if chapter_num <= kapitel_anzahl * 0.25 else ('Hauptteil' if chapter_num <= kapitel_anzahl * 0.75 else 'Schluss')
    current_section_content = roadmap_sections[section]

    role_description = f"""Die Hauptintention dieses Kapitels ist: {intention}. Du arbeitest an Kapitel {chapter_num} von insgesamt {kapitel_anzahl} Kapiteln. Du befindest dich im {section} des Buches. Deine Aufgabe ist es, eine detaillierte Übersicht nur für dieses Kapitel zu erstellen, basierend auf der Roadmap und den Schlüsselmomenten. Beachte den vorherigen Kapitelinhalt und sorge für Kontinuität und Kohärenz. Achte darauf, dass das Kapitel nicht zu "episch" wird und sich auf schrittweise Entwicklungen konzentriert. Deine Antwort sollte eine klare Struktur haben, aber nicht überladen sein. Konzentriere dich auf die Feinheiten und die schrittweise Entwicklung der Geschichte."""

    messages = [
        {"role": "system", "content": f"{role_description}\n\n{buch_prompt}\n\n{current_section_content}\n\nSchlüsselmomente:\n{key_moments}"}
    ]

    if feedback:
        messages.append({"role": "user", "content": feedback})

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-16k",
        messages=messages
    )
    return response.choices[0].message['content'].strip()


# Erstellen Sie einen Ordner "Kapitel-Teile", falls er nicht existiert
if not os.path.exists("Kapitel-Teile"):
    os.makedirs("Kapitel-Teile")

# Aktualisierung der Zusammenfassung
chapter_summaries = []

all_intentions = generate_all_chapter_intentions()

# Feedback-Loop
for chapter_num in range(1, kapitel_anzahl + 1):
    retries = 0
    intention = all_intentions[chapter_num - 1]
    chapter_content = generate_chapter_based_on_intention(chapter_num, intention)
    feedback_messages = [
        f"Der vorherige Versuch passte nicht zur beabsichtigten Intention: '{intention}'. Bitte versuche es erneut und konzentriere dich stärker auf die gewünschte Richtung.",
        f"Es fehlte immer noch der Fokus auf: '{intention}'. Bitte beziehe dich direkt auf diese Absicht und stelle sicher, dass das Kapitel sich darauf konzentriert.",
        f"Das Kapitel sollte sich hauptsächlich um die Intention: '{intention}' drehen. Bitte überarbeite es entsprechend."
    ]

    while not check_coherence(chapter_content, intention) and retries < MAX_RETRIES:
        feedback = feedback_messages[retries] if retries < len(feedback_messages) else f"Bitte beziehe dich direkt auf die Intention: '{intention}'."
        chapter_content = generate_chapter_based_on_intention(chapter_num, intention, feedback)
        retries += 1

    chapter_summaries.append(chapter_content.split('.')[0])  # Speichere den ersten Satz als Zusammenfassung
    with open(f"Kapitel-Teile/Kapitel_{chapter_num}.txt", 'w') as f:
        f.write(chapter_content)
```
